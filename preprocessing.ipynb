```python
import pandas as pd
```

```python
df = pd.read_csv('online_shoppers_intention.csv')
df.head()
```

```python
df.info()
```

```python
# First drop duplicates
df.drop_duplicates(keep='first', inplace=True, ignore_index=True)
df
```

```python
# There are rows where Administrative, Informational, and ProductRelated are all 0
# I don't think this makes sense, so I just remove them

cond1 = df['Administrative'] != 0
cond2 = df['Informational'] != 0
cond3 = df['ProductRelated'] != 0

df = df[cond1 | cond2 | cond3]
df.reset_index(inplace=True, drop=True)
df
```

```python
# There are instances where an Administrative, Informational, or ProductRelated page was visited,
# but the corresponding duration was 0, so I thought just set it to 1 (second)

for x in ['Administrative', 'Informational', 'ProductRelated']:
    df[x+'_Duration'] = pd.Series([
        1 if df[x][i] >= 1 and df[x+'_Duration'][i] == 0 else df[x+'_Duration'][i] for i in range(len(df))
    ])


df
```

```python
# Changing the month into their numerical versions
df.replace({
    'Feb': 2,
    'Mar': 3,
    'May': 5,
    'June': 6,
    'Jul': 7,
    'Aug': 8,
    'Sep': 9,
    'Oct': 10,
    'Nov': 11,
    'Dec': 12
}, inplace=True)

df
```

```python
# One-hot encoding for visitor type to avoid arbitrarily giving numerical values for each type

df = pd.get_dummies(df)
df
```

```python
# Changing Weekend and Revenue into ints

df = df.astype({'Weekend': 'int64', 'Revenue': 'int64'})
df
```

```python
# It might be better if the durations for each type of page were viewed in relation to each other,
# so I thought we could get the proportion of the total time a user spends on each type of page

df['Administrative_Proportion'] = df['Administrative_Duration']/(df['Administrative_Duration']+df['Informational_Duration']+df['ProductRelated_Duration'])
df['Informational_Proportion'] = df['Informational_Duration']/(df['Administrative_Duration']+df['Informational_Duration']+df['ProductRelated_Duration'])
df['ProductRelated_Proportion'] = df['ProductRelated_Duration']/(df['Administrative_Duration']+df['Informational_Duration']+df['ProductRelated_Duration'])
df
```

```python
df.to_csv('cleaned_dataset.csv', index=False)
```
